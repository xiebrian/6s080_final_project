```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_us_2018 = pd.read_csv("report_us_2018.csv")[["Week", "ALL_INF"]]
df_us_2018.columns = ["Week", "Influenza Positive Cases"]
df_aus_2018 = pd.read_csv("report_aus_2018.csv")[["Week", "ALL_INF"]]
df_aus_2018.columns = ["Week", "Influenza Positive Cases"]

df_us_2017 = pd.read_csv("report_us_2017.csv")[["Week", "ALL_INF"]]
df_us_2017.columns = ["Week", "Influenza Positive Cases"]
df_aus_2017 = pd.read_csv("report_aus_2017.csv")[["Week", "ALL_INF"]]
df_aus_2017.columns = ["Week", "Influenza Positive Cases"]
```

```python
df_concat_2018 = pd.concat([df_us_2018.assign(dataset='USA'), df_aus_2018.assign(dataset='AUS')])
df_concat_2017 = pd.concat([df_us_2017.assign(dataset='USA'), df_aus_2017.assign(dataset='AUS')])
```

```python
fig, ax = plt.subplots()
ax.set(yscale="log")

sns.lineplot(x="Week", y="Influenza Positive Cases", data=df_concat_2018, hue="dataset", ax=ax).set_title("Positive cases of influenza in 2018 (by week)")
```

```python
fig, ax = plt.subplots()
ax.set(yscale="log")

sns.lineplot(x="Week", y="Influenza Positive Cases", data=df_concat_2017, hue="dataset", ax=ax).set_title("Positive cases of influenza in 2017 (by week)")
```
